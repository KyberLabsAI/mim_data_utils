```python
from mim_data_utils.logger import BinaryWebSocketServer

import ormsgpack
import os
```

```python
server = BinaryWebSocketServer()
server.start()
```

```python
IMAGE_FILENAME = "test_image.jpg" 

def read_binary_file(filepath: str) -> bytes:
    """Reads a file in binary mode and returns the raw bytes."""
    print(f"Reading file: {filepath}")
    with open(filepath, 'rb') as f:
        # Read the entire file content as a 'bytes' object
        file_data = f.read()
    print(f"File size: {len(file_data)} bytes")
    return file_data

def pack_image_to_msgpack(image_bytes: bytes) -> bytes:
    """
    Packs the image bytes into the desired MessagePack format:
    {'picture': binary_data}
    
    Python's 'bytes' type is natively mapped to MessagePack's 'bin' format, 
    which JavaScript decoders will interpret as a Uint8Array.
    """
    data_to_pack = {
        "picture": image_bytes
    }
    
    packed_message = ormsgpack.packb(data_to_pack)
    
    print(f"MessagePack packed size: {len(packed_message)} bytes")
    return packed_message
```

```python
jpeg_bytes = read_binary_file(IMAGE_FILENAME)

msgpack_data = pack_image_to_msgpack(jpeg_bytes)

server.broadcast(msgpack_data)
```
